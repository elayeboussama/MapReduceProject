## Step 1: Data Exploration

```python
import pandas as pd
import numpy as np

# Charger le jeu de données
df = pd.read_csv('stock_details_5_years.csv')

# Afficher les premières lignes du jeu de données
print(df.head())

# Vérifier les valeurs manquantes
print(df.isnull().sum())

# Statistiques de base
print(df.describe())
```

## Step 2: Data Cleaning

```python
# Supprimer les lignes avec des valeurs manquantes
df = df.dropna()

# Vérifier et supprimer les doublons
df = df.drop_duplicates()

# Convertir la colonne de date au format datetime
df['Date'] = pd.to_datetime(df['Date'])

# Assurer que les types de données sont appropriés
df['Volume'] = df['Volume'].astype(int)

# Sauvegarder le jeu de données nettoyé
df.to_csv('cleaned_dataset.csv', index=False)
```

## Step 3: Data Visualization

```python
import matplotlib.pyplot as plt

# Visualiser les prix des actions pour une entreprise spécifique
company_data = df[df['Company'] == 'AAPL']  # Remplacer 'AAPL' par le symbole de l'entreprise souhaitée
plt.plot(company_data['Date'], company_data['Close'])
plt.title('Prix des actions au fil du temps')
plt.xlabel('Date')
plt.ylabel('Prix de clôture')
plt.show()
```
